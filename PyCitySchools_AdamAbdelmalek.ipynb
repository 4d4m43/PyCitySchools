```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#total number of schools and students using nunique() on school_data_complete dataFrame
total_num_of_schools = school_data_complete["school_name"].nunique()
total_num_of_students = school_data_complete["Student ID"].nunique()

#get list of unique() budget values
total_budget_list = school_data_complete["budget"].unique()
#sum() of list of budget values
total_budget = total_budget_list.sum()

#calculate avg math and reading scores using mean() on columns in dataframe
avg_math_score = school_data_complete["math_score"].mean()
avg_reading_score = school_data_complete["reading_score"].mean()

#insert overall passing rate into data frame, staged with reading score values, then converted to float datatype
school_data_complete['overall_passing_rate'] = school_data_complete['reading_score']
school_data_complete['overall_passing_rate'] = school_data_complete.overall_passing_rate.astype('float64')

#iterate through all students and calculate overall passing rate and insert into dataFrame using at()
i = 0
for i in range(total_num_of_students-1):
    add_num = int(school_data_complete.iloc[i]["reading_score"]) + int(school_data_complete.iloc[i]["math_score"])
    div_num = float(add_num/2)
    school_data_complete.at[i,'overall_passing_rate'] = div_num

#calculate mean() of overall passing rate
avg_passing_score = school_data_complete["overall_passing_rate"].mean()

#passing_math df using conditional statement of math score  >= 70
passing_math = school_data_complete[(school_data_complete['math_score'] >= 70)]

#count() to find number of students passing math
num_passing_math = passing_math['Student ID'].count()

#percent calc for passng_math
percent_passing_math = round(num_passing_math/total_num_of_students*100,2)

#similar calculations except on reading_score column
passing_reading = school_data_complete[(school_data_complete['reading_score'] >= 70)]
num_passing_reading = passing_reading['Student ID'].count()
percent_passing_reading = round(num_passing_reading/total_num_of_students*100,2)

#generate two lists, one for the data and one for the columns, seemed easier than making a dictionary
list_data = [total_num_of_schools,total_num_of_students,total_budget,
            avg_math_score, avg_reading_score, avg_passing_score,
            percent_passing_math, percent_passing_reading]
list_columns = ["total_number_of_schools", "total_number_of_students", "total_budget",
             "average_math_score", "average_reading_score", "average_overall_score", 
            "percent_passing_math", "percent_passing_reading"]

#final_df to present calculated data, single row dataFrame, so used loc() and head() to output results
final_df = pd.DataFrame(columns=list_columns)
final_df.loc[0] = list_data
final_df.head(1)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#generate list of school
list_schools = school_data_complete["school_name"].unique()

#create new data frame
new_df = pd.DataFrame

#generate new list of columns
list_columns = ["school_name", "school_type", "total_students",
             "total_school_budget", "per_student_budget", "average_math_score", 
            "average_reading_score", "percent_passing_math", "percent_passing_reading",
               "overall_passing_rate"]

#add columns to dataframe
new_df = pd.DataFrame(columns=list_columns)
#new_df.head()

#iterate through each school in the list of schools and perform calcs to generate list of results
#generate new dataframe school_df to look at school specific data
#use loc() to output list_results to dataFrame for each school
#list_results lines up with list_columns, again, didnt use dictionary
i = 0
for school in list_schools:
    school_name = school
    school_df = school_data_complete.loc[school_data_complete["school_name"] == school]
    #another way to get the school type but came out as list datatype, otherway gives it to me as string immediately
    #school_type = str(school_df['type'].unique()) 
    school_df = school_df.reset_index(drop=True)
    school_type = school_df.at[0,'type']
    total_students = school_df['Student ID'].nunique()
    total_school_budget = int(school_df['budget'].unique())
    per_student_budget = total_school_budget / total_students
    average_math_score = school_df["math_score"].mean()
    average_reading_score = school_df["reading_score"].mean()
    average_overall_score = (average_math_score + average_reading_score)/2
    
    passing_math_school = school_df[(school_df['math_score'] >= 70)]
    num_passing_math_school = passing_math_school['Student ID'].count()
    percent_passing_math_school = round(num_passing_math_school/total_students*100,2)

    passing_reading_school = school_df[(school_df['reading_score'] >= 70)]
    num_passing_reading_school = passing_reading_school['Student ID'].count()
    percent_passing_reading_school = round(num_passing_reading_school/total_students*100,2)
    
    average_overall_score_school = (percent_passing_math_school + percent_passing_reading_school)/2
    
    list_results = [school_name,school_type,total_students,
                   total_school_budget,per_student_budget,average_math_score,
                   average_reading_score,percent_passing_math_school,percent_passing_reading_school,
                   average_overall_score_school]
    new_df.loc[i] = list_results
    i += 1
    
#output results
new_df   
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#get list of unique grade values
list_grades = school_data_complete["grade"].unique()

#create math and reading dataFrames and set columns
math_df = pd.DataFrame(columns=list_grades)
reading_df = pd.DataFrame(columns=list_grades)

#iterate through each school and then each grade, perform calcs and output to appropriate dataFrame
i = 0
for school in list_schools:
    j = 0
    for grade in list_grades:
        #set correct dataFrame by getting pertinent data based on grade and schoolname
        frame = school_data_complete[(school_data_complete['grade'] == grade) & (school_data_complete['school_name'] == school)]
        #calculate mean()
        average_math_score = frame["math_score"].mean()
        average_reading_score = frame["reading_score"].mean()
        #insert values into dataframes
        math_df.at[school, grade] = average_math_score
        reading_df.at[school, grade] = average_reading_score           
        j = j + 1

#output results of math dataframe
math_df
```
